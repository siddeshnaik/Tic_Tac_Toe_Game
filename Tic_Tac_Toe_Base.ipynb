## Game Code

```python
import random
game_mat = [
    ['_', '_', '_'],
    ['_', '_', '_'],
    ['_', '_', '_']
]

#game display

def game_display(game_mat):
    print('Game')
    print('', 0, 1, 2,sep='\t')
    for i, value in enumerate(game_mat):
        print(i,value[0], value[1], value[2], sep='\t')

        
#Game Instances   
class GamePlay:
    def __init__(self, player_symbol,game_mode="CompVsPlayer", player_name='Computer'):
        self.player_name=player_name
        self.game_mode = game_mode
        self.player_symbol = player_symbol
        
    
    def make_move(self, game_mat_instance):
        move_loop = True
        while move_loop:
            if ((self.player_name == 'Computer') and (self.game_mode =='CompVsPlayer')): #Computer making move
                self.move_x = random.randint(0,2)
                self.move_y = random.randint(0,2)
                #print(move_x, move_y)

                if (game_mat_instance[self.move_x][self.move_y] =='_'):
                    move_loop=False

                else:
                    continue

            else:
                try:
                    print('\n')
                    print(self.player_name)
                    print('Play move {}'.format(self.player_symbol))
                    self.move_x = int(input("Enter Location X Co-ordinate: "))
                    self.move_y = int(input("Enter Location Y Co-ordinate: "))

                    if((self.move_x >= 0 and self.move_y >= 0) and (self.move_x <= 2 and self.move_y <= 2)): #valid integers in range 0 to 2
                        if(game_mat_instance[self.move_x][self.move_y]=='_'):
                            move_loop=False

                        else:
                            print('Position is already occupied or Input location out of Bounds')
                            continue


                except ValueError as e:
                    print("Valid Position Co-ordinates have to be integers")
                    
            
        return (self.move_x, self.move_y)
    
        

# move made check and victory check

def move_victory_check(player_name, player_move, player_symbol, game_mat):
    game_mat[player_move[0]][player_move[1]]=player_symbol
    
    #game display code
    print('\n')
    print(player_name)
    print('Game')
    print('', 0, 1, 2,sep='\t')
    for i, value in enumerate(game_mat):
        print(i,value[0], value[1], value[2], sep='\t')
        
    #Victory checks
    outcome = 'Continue'
    
    #vertical win across columns
    for value in game_mat:
        if ((value[0] == value[1] == value[2]) and value[0]!='_'):
            print('\n{}: Victory'.format(player_name))
            outcome = 'Victory'
            break
    
    #diagonal validation/victory
    if ((game_mat[0][0]==game_mat[1][1]==game_mat[2][2]!='_') or (game_mat[2][0]==game_mat[1][1]==game_mat[0][2]!='_')):
        print('\n{}: Victory'.format(player_name))
        outcome = 'Victory'
        
    # horizontal or row wise check
    for i in range(3):
        if ((game_mat[0][i]==game_mat[1][i]==game_mat[2][i]) and game_mat[0][i]!='_'):
            print('\n{}: Victory'.format(player_name))
            outcome = 'Victory'
            break
            
    return (game_mat, outcome)   

#Game mode input
while True:
    try:
        game_mode_value = int(input('Enter Game Mode: \n 0-Computer V/s Player \n 1-Player V/s Player \n :'))
        if (game_mode_value > 1) | (game_mode_value < 0):
            print('Enter a valid Game Mode \n')
            continue
        
        else:
            break
        
    except ValueError as e:
        print('Enter a valid Game Mode as a number \n')

#game mode string matching
mode_dict={0: "CompVsPlayer", 1: 'PlayerVsPlayer'}
game_mode = mode_dict[game_mode_value]
game_display(game_mat)

if (game_mode == "CompVsPlayer"):
    player_name=input('Enter Player name: ')
    while True:
        player_symbol = input('X or O in Upper Case ONLY!: ')
        if (player_symbol=='X'):
            player2_symbol = 'O'
            player2_name = 'Computer'
            player1_name = player_name
            player1_symbol = player_symbol
            break
        
        elif (player_symbol=='O'):
            player1_symbol = 'X'
            player1_name = 'Computer'
            player2_name = player_name
            player2_symbol = player_symbol
            break
        
        else:
            print('Enter X or O only')
            
    # p1 will always go first 
    p1 = GamePlay(game_mode="CompVsPlayer", player_name=player1_name, player_symbol=player1_symbol)
    p2 = GamePlay(game_mode="CompVsPlayer", player_name=player2_name, player_symbol=player2_symbol)
    
else:
    player1_name = input('Enter Player 1 name: ')
    player2_name = input('Enter Player 2 name: ')
    
    p1 = GamePlay(game_mode='PlayerVsPlayer', player_name=player1_name, player_symbol="X")
    p2 = GamePlay(game_mode='PlayerVsPlayer', player_name=player2_name, player_symbol="O")
    
#player initialization

players=[p1,p2]
move_outcome = 'Continue'
count=0
for i in range(9):
    if move_outcome == 'Continue':
        game_mat, move_outcome= move_victory_check(players[i%2].player_name, players[i%2].make_move(game_mat), players[i%2].player_symbol, game_mat)
        count=count+1
        
    else:
        break

if (count==8): # game draw
    print('Game Draw')
```
